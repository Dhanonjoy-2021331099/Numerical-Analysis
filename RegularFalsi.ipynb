```python
import numpy as np
import matplotlib.pyplot as plt

# Function define
def f(x):
    return x**3 - 4*x - 9

# Initial guesses (must satisfy f(a)*f(b) < 0)
a = 2
b = 3
tol = 1e-5
max_iter = 50

if f(a)*f(b) >= 0:
    print("Invalid interval. f(a)*f(b) must be < 0.")
else:
    print("Iter |     a     |     b     |     c     |    f(a)   |    f(b)   |    f(c)   |   ARE (%)")
    print("-------------------------------------------------------------------------------------------")

    prev_c = None

    for i in range(1, max_iter+1):
        c = ( a*f(b) - b*f(a) ) / (f(b) - f(a))
        f_a = f(a)
        f_b = f(b)
        f_c = f(c)

        # Absolute Relative Error
        if prev_c is None:
            are = "-"
        else:
            are = abs((c - prev_c) / c) * 100

        print(f"{i:4d} | {a:9.6f} | {b:9.6f} | {c:9.6f} | {f_a:9.6f} | {f_b:9.6f} | {f_c:9.6f} | {are if are=='-' else f'{are:9.6f}'}")

        # Check stopping condition
        if abs(f_c) < tol:
            print(f"\n Root found at x = {c:.6f} after {i} iterations.")
            break

        # Update interval
        if f_a * f_c < 0:
            b = c
        else:
            a = c

        prev_c = c


```
