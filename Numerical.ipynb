```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 4, 6, 5])
y = np.array([0,1.386294,1.791759,1.609438])

co = np.polyfit(x, y, 3)
poly = np.poly1d(co)
print("The polynomial equation is:")
print(poly)

# # Verify the fit by calculating y for each x and comparing with original y
# for i in range(len(x)):
#     y_calculated = polynomial(x[i])
#     print(f"At x = {x[i]}, Original y = {y[i]}, Calculated y = {y_calculated}")

# Plotting
x_plot = np.linspace(min(x), max(x), 500)
y_plot = poly(x_plot)

plt.scatter(x, y, color='red')
plt.plot(x_plot, y_plot,color='blue')
plt.title('Polynomial Fit')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()
plt.show()
```
