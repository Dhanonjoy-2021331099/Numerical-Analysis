```python
import matplotlib.pyplot as pt
import numpy as np
import pandas as pd
```

```python
# CREATE POLYNOMIAL fUNCTION

x = np.array([1,5,8])
y = np.array([3,7,0])

length = x.size
co = np.polyfit(x,y,length-1)

poly =np.poly1d(co)
print('Ploynomial Function is : ')
print(poly)
```

```python
# CURVE PLOT
xplot = np.linspace(min(x),max(x),5000)
yplot = poly(xplot)

pt.plot(xplot,yplot,label='Polynomail Fit')
pt.scatter(x,y,color='red',label='Data Points')
pt.legend()
pt.show()
```
